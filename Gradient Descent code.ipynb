```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0  # m_curr = theta1, b_curr = theta0
    iterations = 100   # change according to requirement
    n = len(x)  # length of x and y should be same
    learning_rate =  0.01  # learning_rate = alpha, change according to requirement
    
    for i in range(iterations):
        y_predicted = (m_curr * x) + b_curr  # hypothesis h(x) = theta0 + theta1 * x = y_predicted
        cost_func = (1/n) * sum([val**2 for val in (y-y_predicted)])   #Cost function J(theta0, theta1), this has to reduce for good gradient descent
        md = -(2/n) * sum(x * (y-y_predicted))
        bd = -(2/n) * sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m: {}, b: {}, cost function: {}, iteration: {}".format(m_curr, b_curr, cost_func, i))
```

```python
# Example 
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
gradient_descent(x,y)
```
